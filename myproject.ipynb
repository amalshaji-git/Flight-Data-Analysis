13. A PLOT OF THE FIRST 50 FLIGHTS DEPARTURE AND ARRIVAL DELAYS

```python
import matplotlib.pyplot as plt
# Selecting first 50 departure and arrival delays
first_50_dep_delays = flight['dep_delay'].head(50)
first_50_arr_delays = flight['arr_delay'].head(50)
# Create a figure with specified size
plt.figure(figsize=(12, 6))
# Plotting departure delays as blue circles and arrival delays as red crosses
plt.plot(first_50_dep_delays, marker='o', linestyle='-', color='b', label='Departure Delay')
plt.plot(first_50_arr_delays, marker='x', linestyle='-', color='r', label='Arrival Delay')
# Adding title and axis labels
plt.title('First 50 Flight Delay Times')
plt.xlabel('Flight Index')
plt.ylabel('Delay in Minutes')
# Adding legend, grid, and displaying the plot
plt.legend()
plt.grid(True)
plt.show()
```

14. A HEATMAP OF FLIGHT DELAYS BY HOUR AND DAY

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Convert sched_dep_time to hour and day_of_week
flight['hour'] = pd.to_datetime(flight['sched_dep_time'], format='%H%M').dt.hour
flight['day_of_week'] = pd.to_datetime(flight[['year', 'month', 'day']]).dt.dayofweek
# Create a pivot table for heatmap data
heatmap_data = flight.pivot_table(values='dep_delay', index='day_of_week', columns='hour', aggfunc='mean')
# Create a figure with specified size
plt.figure(figsize=(12, 6))
# Create heatmap using Seaborn
sns.heatmap(heatmap_data, cmap='coolwarm', annot=True)
# Set plot title and axis labels
plt.title('Heatmap of Average Departure Delays by Hour and Day')
plt.xlabel('Hour of Day')
plt.ylabel('Day of Week')
# Display the plot
plt.show()
```

It is  a heatmap showing average flight departure delays by hour and day of the week using a pivot table, with color intensity representing the delay magnitude and annotations displaying the average delays.

18. Bar Chart for Flights Delayed by More Than 60 Minutes at Departure and Also Delayed at Arrival

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Filter flights delayed more than 60 minutes at departure and also delayed at arrival
delayed_flights = flight[(flight['dep_delay'] > 60) & (flight['arr_delay'] > 0)]

# Plot the number of delayed flights by carrier
sns.countplot(data=delayed_flights, x='carrier')
plt.title('Number of Flights Delayed > 60 Minutes at Departure and Also Delayed at Arrival by Carrier')
plt.xlabel('Carrier')
plt.ylabel('Number of Flights')
plt.xticks(rotation=45)
plt.show()
```

 19. Estimating and visualize the total revenue generated by each flight using the distance traveled (BAR CHART)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Define a rate per mile (assuming a rate of $0.10 per mile)
rate_per_mile = 0.10
# Calculate revenue for each flight based on distance
flight['revenue'] = flight['distance'] * rate_per_mile
# Group by flight name and sum up the revenue
revenue_by_flight = flight.groupby('name')['revenue'].sum().reset_index()
# Sort by revenue to make the bar chart more informative
revenue_by_flight = revenue_by_flight.sort_values(by='revenue', ascending=False)
# Plot the revenue for each flight name
plt.figure(figsize=(10, 6))
# Use hue with x variable and disable legend
sns.barplot(data=revenue_by_flight, x='name', y='revenue', hue='name', palette='viridis', dodge=False)
plt.title('Estimated Revenue by Flight Name')
plt.xlabel('Flight Name')
plt.ylabel('Estimated Revenue ($)')
plt.xticks(rotation=90)
plt.legend([], [], frameon=False)  # Remove legend
plt.show()
```

This code calculates the estimated revenue for each flight in a dataset based on the distance traveled, using a fixed rate per mile (e.g., $0.10 per mile). The revenue is then summed up for each flight name, sorted in descending order, and visualized using a bar chart. The visualization helps identify which flights are generating the most revenue, allowing for easier comparison and potential insights into optimizing flight operations.

20. Estimating and visualize the total revenue generated by each flight using the distance traveled (Scatter Plot)

```python
import pandas as pd
import matplotlib.pyplot as plt
# Define a rate per mile (assuming a rate of $0.10 per mile)
rate_per_mile = 0.10
# Calculate revenue for each flight based on distance
flight['revenue'] = flight['distance'] * rate_per_mile
# Group by flight name and sum up the revenue
revenue_by_flight = flight.groupby('name')['revenue'].sum().reset_index()
# Sort by revenue to make the plot more informative
revenue_by_flight = revenue_by_flight.sort_values(by='revenue', ascending=False)
# Plot the revenue for each flight name using a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(revenue_by_flight['revenue'], revenue_by_flight['name'], color='green')
plt.title('Estimated Revenue by Flight Name')
plt.xlabel('Estimated Revenue ($)')
plt.ylabel('Flight Name')
plt.show()
```

The scatter plot places the flight names on the y-axis and the corresponding revenue on the x-axis. This plot is helpful for quickly identifying outliers or flights with significantly higher or lower revenue.

21. Visualization of Departure and Arrival Delays for Flights Departing Late but Arriving on Time

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Filter flights that departed late but did not arrive late
late_departure_no_late_arrival = flight[(flight['dep_delay'] > 0) & (flight['arr_delay'] <= 0)]
# Select the first 10 flights
first_100_flights = late_departure_no_late_arrival.head(100)
# Create the figure and axes for the plot
plt.figure(figsize=(26, 8))
# Create a stacked bar chart
bar_width = 0.5
index = np.arange(len(first_100_flights))
# Bar for departure delay
plt.bar(index, first_100_flights['dep_delay'], bar_width, color='blue', label='Departure Delay')
# Bar for arrival delay (stacked on top of departure delay)
plt.bar(index, first_100_flights['arr_delay'], bar_width, bottom=first_100_flights['dep_delay'], color='green', label='Arrival Delay')
# Add labels and title
plt.title('Comparison of Departure and Arrival Delays for First 10 Flights')
plt.xlabel('Flight Number')
plt.ylabel('Delay (Minutes)')
plt.xticks(index, first_100_flights['flight'], rotation=90)
plt.legend()
# Show the plot
plt.show()
```

This script analyzes flight data to identify flights that departed late but did not arrive late. It filters such flights from the dataset and selects the first 100 flights for visualization. A stacked bar chart is then created to compare the departure delays and arrival delays for these flights. The x-axis represents the flight numbers, while the y-axis shows the delay in minutes. Departure delays are shown in blue, and arrival delays (stacked on top) are shown in green. The chart provides a clear visual comparison, highlighting the relationship between departure and arrival delays for these specific flights.
